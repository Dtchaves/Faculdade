```python
size_table = SPA_DF['HT'].unique().size
table_esp = pd.DataFrame(SPA_DF['HT'].unique(), columns=['Times'])
table_esp["Nº de pontos"] = np.zeros(size_table, dtype=int)
table_esp["Jogos"] = np.zeros(size_table, dtype=int)
table_esp["Vitórias"] = np.zeros(size_table, dtype=int)
table_esp["Empates"] = np.zeros(size_table, dtype=int)
table_esp["Derrotas"] = np.zeros(size_table, dtype=int)
table_esp["Gols pró"] = np.zeros(size_table, dtype=int)
table_esp["Gols contra"] = np.zeros(size_table, dtype=int)
table_esp["Saldo de gols"] = np.zeros(size_table, dtype=int)

for index, row in SPA_DF.iterrows():
    
    index_HT = table_esp['Times'] == row['HT']
    index_AT = table_esp['Times'] == row['AT']
    
    table_esp.loc[index_HT, 'Jogos'] += 1
    table_esp.loc[index_AT, 'Jogos'] += 1

    if row['WDL'] == 'W':
        table_esp.loc[index_HT, 'Vitórias'] += 1
        table_esp.loc[index_AT, 'Derrotas'] += 1
        
    elif row['WDL'] == 'L':
        table_esp.loc[index_HT, 'Derrotas'] += 1
        table_esp.loc[index_AT, 'Vitórias'] += 1
        
    elif row['WDL'] == 'D':
        table_esp.loc[index_HT, 'Empates'] += 1
        table_esp.loc[index_AT, 'Empates'] += 1
    
    table_esp.loc[index_HT, 'Gols pró'] += row['HS']
    table_esp.loc[index_HT, 'Gols contra'] += row['AS']

    
    table_esp.loc[index_AT, 'Gols pró'] += row['AS']
    table_esp.loc[index_AT, 'Gols contra'] += row['HS']
    
table_esp['Nº de pontos'] = 3*table_esp['Vitórias'] + table_esp['Empates']
table_esp['Saldo de gols'] = table_esp['Gols pró'] - table_esp['Gols contra']
table_spa = df.sort_values(by=['Nº de pontos', 'Vitórias', 'Saldo de gols', 'Gols pró'], 
                             ascending=[False, False, False, False]).reset_index(drop=True)
```

## Questão 4
- Utilizando os jogos da liga escolhida, use regressão de Poisson para criar um modelo de previsão de resultados, como visto nos slides em sala e no Soccermatics.
-- https://soccermatics.readthedocs.io/en/latest/gallery/lesson5/plot_SimulateMatches.html
- Dê print no sumário do ajuste
- Simule a partida entre o 1º e o 4º colocado, onde o 1º joga em casa. Primeiro, apresente a quantidade esperada de gols de cada time. Em seguida, apresente um grid com as probabilidades associadas a cada placar possível.

```python
goal_model_data = pd.concat([SPA_DF[['HT','AT','HS']].assign(home=1).rename(
            columns={'HT':'team', 'AT':'opponent','HS':'goals'}),
           SPA_DF[['HT','AT','HS']].assign(home=0).rename(
            columns={'AwayTeam':'team', 'HomeTeam':'opponent','AS':'goals'})])

poisson_model = smf.glm(formula="goals ~ home + team + opponent", data=goal_model_data,
                        family=sm.families.Poisson()).fit()
print(poisson_model.summary())
```

```python
home_team='Real Madrid'
away_team='Sevilla FC'

home_score_rate=poisson_model.predict(pd.DataFrame(data={'team': home_team, 'opponent': away_team,
                                       'home':1},index=[1]))
away_score_rate=poisson_model.predict(pd.DataFrame(data={'team': away_team, 'opponent': home_team,
                                       'home':0},index=[1]))
print(home_team + ' against ' + away_team + ' expect to score: ' + str(home_score_rate))
print(away_team + ' against ' + home_team + ' expect to score: ' + str(away_score_rate))

home_goals=np.random.poisson(home_score_rate)
away_goals=np.random.poisson(away_score_rate)
print(home_team + ': ' + str(home_goals[0]))
print(away_team + ': '  + str(away_goals[0]))
```

```python

max_goals=5

team_pred = [[poisson.pmf(i, team_avg) for i in range(0, max_goals + 1)] for team_avg in
                 [home_score_rate, away_score_rate]]

score_matrix=np.outer(np.array(team_pred[0]), np.array(team_pred[1]))

fig=plt.figure()
ax=fig.add_subplot(1,1,1)
pos=ax.imshow(score_matrix, extent=[-0.5,max_goals+0.5,-0.5,max_goals+0.5], aspect='auto',cmap=plt.cm.Blues)
fig.colorbar(pos, ax=ax)
ax.set_title('Probability of outcome')
plt.xlim((-0.5,5.5))
plt.ylim((-0.5,5.5))
plt.tight_layout()
ax.set_xlabel('Goals scored by ' + away_team)
ax.set_ylabel('Goals scored by ' + home_team)
plt.show()

homewin=np.sum(np.tril(score_matrix, -1))
draw=np.sum(np.diag(score_matrix))
awaywin=np.sum(np.triu(score_matrix, 1))
```

```python
size = SPA_DF.shape[0]
SPA_DF_SIM = pd.DataFrame(columns=["HT","AT","HS","AS","WDL"])
```

```python
for home_team in SPA_DF['HT'].unique():
    for away_team in SPA_DF['AT'].unique():
        
        if home_team == away_team:
            continue 
            
        home_score_rate=poisson_model.predict(pd.DataFrame(data={'team': home_team, 'opponent': away_team,
                                        'home':1},index=[1]))

        away_score_rate=poisson_model.predict(pd.DataFrame(data={'team': away_team, 'opponent': home_team,
                                        'home':0},index=[1]))
                
        home_goals=np.random.poisson(home_score_rate)
        away_goals=np.random.poisson(away_score_rate)
        
        if home_goals > away_goals:
            wdl = 'W'
        elif home_goals < away_goals:
            wdl = 'L'
        elif home_goals == away_goals:
            wdl = 'D'
        
        game =  {"HT":home_team,"AT":away_team,"HS":int(home_goals[0]),"AS":int(away_goals[0]),"WDL":wdl}
        SPA_DF_SIM.loc[len(SPA_DF_SIM)] = game
```
