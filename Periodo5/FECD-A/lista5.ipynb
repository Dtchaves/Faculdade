```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dados da tabela
ranks = np.array([1, 2, 27, 802, 2087, 8901, 14343, 21531])  # preencha com os dados fornecidos
frequencies = np.array([79607, 48238, 4033, 174, 70, 12, 6, 3])  # preencha com os dados fornecidos

# Cálculo dos logaritmos
log_ranks = np.log(ranks)
log_frequencies = np.log(frequencies)

# Regressão linear
model = LinearRegression()
model.fit(log_ranks.reshape(-1, 1), log_frequencies)

# Coeficientes da linha ajustada
slope = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(log_ranks.reshape(-1, 1), log_frequencies)

# Plotagem do gráfico
plt.scatter(log_ranks, log_frequencies,color='black')
plt.plot(log_ranks, model.predict(log_ranks.reshape(-1, 1)), color='red')
plt.xlabel('Rank')
plt.grid(True)
plt.ylabel('Frequência')
plt.title('Ajuste Linear')
plt.show()

print(f"Slope (theta): {-slope}")
print(f"Intercept (C): {np.exp(intercept)}")
print(f"R-squared: {r_squared}")
```

```python
import numpy as np
from scipy.stats import kstest, norm

# Defina os parâmetros para a distribuição normal
mu = 0
sigma = 1
n = 50

# Gere a amostra da distribuição normal
np.random.seed(0)  # para reprodutibilidade
x = np.random.normal(mu, sigma, n)

# Realize o teste K-S
ks_statistic, ks_p_value = kstest(x, 'norm', args=(mu, sigma))

print(f"Statistic: {ks_statistic}, p-value: {ks_p_value}")

# Se necessário, altere mu ou sigma para realizar o teste com um modelo diferente
```

```python
from scipy.stats import chi2

D = np.array([49.78, 83.31, 128.35, -385.65, 331.40, -480.87, 97.76, 98.22, 77.70])
# Valor esperado
E = np.array([165.22, 1401.69, 5202.65, 11034.65, 14627.60, 12409.87, 6580.24, 1993.78, 264.30])# Substitua expected_value1, expected_value2, ... pelos valores esperados

# Calcular a estatística qui-quadrado
X2 = sum((d)**2 / e for d, e in zip(D, E))

print("Estatística Qui-Quadrado:", X2)

# Graus de liberdade (número de categorias - 1)
df = len(E) - 1

# Calcular o p-valor
p_valor = chi2.sf(X2, df)

print("P-Valor:", p_valor)
```
