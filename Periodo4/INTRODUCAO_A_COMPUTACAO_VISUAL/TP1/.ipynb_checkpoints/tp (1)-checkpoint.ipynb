```python
blocksize = 8 
#Divide a imagem em blocos
height, width = camera.shape
num_blocks_height = height // blocksize
num_blocks_width = width // blocksize

# Aplica a DCT em cada bloco
dct_kernels = np.zeros((num_blocks_height, num_blocks_width, blocksize, blocksize))

for i in range(num_blocks_height):
    for j in range(num_blocks_width):
        #Obtem um bloco da imagem
        block = camera[i*blocksize:(i+1)*blocksize, j*blocksize:(j+1)*blocksize]
        #Aplica a DCT no bloco
        dct_block = cv2.dct(np.float32(block))
        dct_kernels[i,j,:,:] = dct_block


quantization_matrix = np.array([[16,11,10,16,24,40,51,61],
                                [12,12,14,19,26,58,60,55],
                                [14,13,16,24,40,57,69,56],
                                [14,17,22,29,51,87,80,62],
                                [18,22,37,56,68,109,103,77],
                                [24,35,55,64,81,104,113,92],
                                [49,64,78,87,103,121,120,101],
                                [72,92,95,98,112,100,103,99]])

#Comprimindo a imagem
for i in range(num_blocks_height):
    for j in range(num_blocks_width):
        dct_block = dct_kernels[i,j,:,:]
        dct_block = np.round(dct_block / quantization_matrix)
        dct_kernels[i,j,:,:] = dct_block

#Revertendo o processo
for i in range(num_blocks_height):
    for j in range(num_blocks_width):
        dct_block = dct_kernels[i,j,:,:]
        dct_block = dct_block * quantization_matrix
        dct_kernels[i,j,:,:] = dct_block

#Reconstruindo a imagem
reconstructed_image = np.zeros((height, width), dtype=np.float32)

for i in range(num_blocks_height):
    for j in range(num_blocks_width):
        block = dct_kernels[i,j,:,:]
        block = cv2.idct(block)
        
        ystart, xstart = i*blocksize, j*blocksize
        yend, xend = ystart+blocksize, xstart+blocksize
        reconstructed_image[ystart:yend, xstart:xend] = block

#Exibe a imagem original e a reconstru√≠da
plt.figure(figsize=(10,10))
plt.subplot(121), plt.imshow(camera, cmap='gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(reconstructed_image, cmap='gray')
plt.title('Imagem Reconstru√≠da'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Aplicando m√©tricas de qualidade 
#RMSE 
from math import sqrt
import skimage.measure as measure
from skimage.metrics import mean_squared_error 
mse = mean_squared_error(camera, reconstructed_image)
print("MSE: ", mse)

#PSNR
psnr = 10 * np.log10((255**2) / mse)
print("PSNR: ", psnr)

```
