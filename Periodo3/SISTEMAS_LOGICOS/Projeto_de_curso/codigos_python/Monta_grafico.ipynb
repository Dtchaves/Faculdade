```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Entrada = pd.read_csv('dados/n=.csv')
```

```python
Entrada["Y"] = Entrada["Y"].apply(lambda x: int(str(x), 2))
```

```python
sns.set(style='ticks', font_scale=1.0, rc={'axes.facecolor': 'white', 'figure.facecolor': 'white'})

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=Entrada, color="blue", linewidth=1, marker='o')


ax.set_xlabel('Iteraes', fontsize=10, fontweight='bold')
ax.set_ylabel('Valor', fontsize=10, fontweight='bold')

plt.title('n = , x + x + 1 \n', fontsize=15)

plt.savefig('resultados/n=.png', dpi=300, bbox_inches='tight')

plt.show()
```
